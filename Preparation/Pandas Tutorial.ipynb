## Loading data into Pandas

```python
import pandas as pd

#read data from csv into a pandas dataframe, optional: define delimiter
df = pd.read_csv('movie_data.csv', delimiter=',')
```

## Reading Data in Pandas

```python
#print first 4 rows of the dataframe
print(df.head(4))

#access column by name
df['Genre']

#easily convert 1-dimensional datas into python lists
df['Genre'].to_list()

#get column names
df.columns

#get subset of the dataframe
df[['Genre', 'Lead Studio', 'Profitability']]

#get specific rows
df.iloc[10:15]

#iterate rows
for index, row in df.iterrows():
    print(row['Film'])

#access all rows where certain condition is met
df.loc[df['Profitability'] > 10]

#access a specific location in the dataframe 
print(df.iloc[5,2])
```

## Sorting Data

```python
#sort data by multiple attributes. example: sort movies by critics score, then audience score
df.sort_values(by=['Rotten Tomatoes %', 'Audience score %'], ascending=[0,0])
```

## Exporting the Dataframe

```python
df.to_csv('new_movie_data.csv', index=False, sep=',')

df.to_excel('new_movie_data.xlsx', index=False)
```

## Filtering Data

```python
#filter data by certain criteria, save results in a new dataframe
new_df = df.loc[(df['Year'] >= 2010) & (df['Genre'] == 'Comedy')]

#compute new index for the new dataframe, do this inplace and not on a new copy
new_df.reset_index(drop=True, inplace=True)

new_df
```

## Group By

```python
#group genres by their mean profitability and sort them
df.groupby(['Genre']).mean(numeric_only=True).sort_values('Profitability', ascending=False)


#count number of entries in a category
df.groupby(['Genre']).count()
```
